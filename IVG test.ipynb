```python
# Initialiser une liste pour stocker les départements avec les valeurs les plus élevées par année
departements_max = []

# Pour chaque colonne année, trouver la zone géographique avec la valeur maximale
for colonne in df2.columns[1:]:  # Commencer à partir de la deuxième colonne (après la colonne 'Zone géographique')
    # Convertir les valeurs en float en remplaçant les virgules par des points et en convertissant en float
    df2[colonne] = df2[colonne].str.replace(',', '.').astype(float)
    # Trouver l'index de la ligne avec la valeur maximale pour cette année
    max_value_index = df2[colonne].idxmax()
    # Extraire la zone géographique correspondante
    zone_geographique_max = df2.loc[max_value_index, 'Zone géographique']
    # Ajouter le résultat à la liste des départements_max
    departements_max.append((colonne, zone_geographique_max))

# Afficher les résultats
for annee, zone_geographique in departements_max:
    print(f"Pour l'année {annee}, la zone géographique avec la valeur maximale est {zone_geographique}.")
```

```python
# Initialiser une liste pour stocker les départements avec les valeurs les plus faibles par année
departements_min = []

# Pour chaque colonne année, trouver la zone géographique avec la valeur minimale
for colonne in df2.columns[1:]:  # Commencer à partir de la deuxième colonne (après la colonne 'Zone géographique')
    # Vérifier si les valeurs sont déjà de type float
    if not pd.api.types.is_numeric_dtype(df2[colonne]):
        # Convertir les valeurs en float en remplaçant les virgules par des points et en convertissant en float
        df2[colonne] = df2[colonne].str.replace(',', '.').astype(float)
    # Trouver l'index de la ligne avec la valeur minimale pour cette année
    min_value_index = df2[colonne].idxmin()
    # Extraire la zone géographique correspondante
    zone_geographique_min = df2.loc[min_value_index, 'Zone géographique']
    # Ajouter le résultat à la liste des départements_min
    departements_min.append((colonne, zone_geographique_min))

# Afficher les résultats
for annee, zone_geographique in departements_min:
    print(f"Pour l'année {annee}, la zone géographique avec la valeur minimale est {zone_geographique}.")
```

```python
import matplotlib.pyplot as plt

# Fonction personnalisée pour afficher la somme au lieu du pourcentage
def sum_formatter(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    return f"{absolute}\n({pct:.1f}%)"

# Couleurs personnalisées pour chaque tranche
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Calculer la somme de chaque colonne année pour obtenir les quantités totales par année
sum_per_year_df1 = df1.drop(columns='département').sum()

# Créer un diagramme circulaire avec des bordures aux tranches
plt.figure(figsize=(8, 8))
plt.pie(sum_per_year_df1, labels=sum_per_year_df1.index, autopct=lambda pct: sum_formatter(pct, sum_per_year_df1), startangle=140, colors=colors, wedgeprops=dict(edgecolor='black', linewidth=1.5))
plt.title('Quantités d\'IVG par année')
plt.axis('equal')  # Assure que le diagramme circulaire est affiché de manière circulaire
plt.show()
```
